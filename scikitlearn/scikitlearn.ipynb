```python
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data=np.loadtxt(os.path.join('Data','shopdata.txt'),delimiter=',')
X=data[:,0]
Y=data[:,1]
m=Y.size
```

```python
print("print data")
def plotData(x,y):
    plt.plot(x,y,'ro',ms=10,mec='k')
    plt.xlabel('population in the city')
    plt.ylabel('profit from the city')
    fig=plt.figure()
```

```python
plotData(X,Y)
```

```python
print(X)
```

```python
X=X.reshape(-1,1)
Y=Y.reshape(-1,1)
```

```python
print(X.shape)
```

```python
print(Y.shape)
```

```python
lr=LinearRegression()
lr.fit(X,Y)
```

```python
y_predict=lr.predict(X)
plt.scatter(X,Y, color='g')
plt.plot(X,y_predict ,color='k')
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
text_x=np.array([[6.23]])
text_y=lr.predict(text_x)
print(text_y)
```

```python
print(lr.score(X,Y))
```

```python
mean_squared_error(Y, y_predict)
```
