```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris_data=pd.read_csv('iris.csv')
print(iris_data)
```

```python
X=iris_data.iloc[:,0:4]
y=iris_data.iloc[:,-1]
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
y_le=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y_le,test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
sc = preprocessing.StandardScaler()
sc.fit(X_train)
X_train_scale=sc.transform(X_train)
X_test_scale=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5 ,p=2,metric='minkowski')
```

```python
knn.fit(X_train_scale,Y_train)
```

```python
y_pred=knn.predict(X_test_scale)
```

```python
from sklearn.metrics import accuracy_score
print('miss classified items are: %d'%(Y_test !=y_pred).sum())
```

```python
print('accuray score:%.2f'% accuracy_score(Y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```

SVM

```python
from sklearn.svm import SVC
svc=SVC(kernel='rbf')
svc.fit(X_train_scale,Y_train)
```

```python
y_pred_svc=svc.predict(X_test)
```

```python
print('miss classified :%d '%(Y_test != y_pred_svc).sum())
```

```python
print('accuray score:%.2f'% accuracy_score(Y_test,y_pred_svc))
```

```python
print(classification_report(Y_test,y_pred_svc))
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(C=1,random_state=0)
lr.fit(X_train_scale,Y_train)
```

```python
y_pred_log=lr.predict(X_test_scale)
print('missmatched:%d'%(Y_test !=y_pred_log).sum())
```

```python
print('accuracy:%.2f'%accuracy_score(Y_test,y_pred_log))
```

```python
print(classification_report(Y_test,y_pred_log))
```
