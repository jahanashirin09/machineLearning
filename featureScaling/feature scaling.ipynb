```python

import numpy as np
import random
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
y=np.random.random(300)
random.seed(1)
x=np.arange(300)
```

```python
print(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
def standadization(X):
    mean=np.mean(X,axis=0)
    std=np.std(X,axis=0)
    X_std=(X-mean)/std
    return X_std
```

```python
Y=standadization(y)
```

```python
fig,(axis1,axis2)=plt.subplots(ncols=2,figsize=(6,5))
axis1.set_title('before standadization')
sns.kdeplot(y,ax=axis1)
axis2.set_title('after standadization')
sns.kdeplot(Y,ax=axis2)
```

```python
def normalization(X):
    Xmin=np.min(X)
    Xmax=np.max(X)
    X_norm=(X-Xmin)/(Xmax-Xmin)
    return X_norm
```

```python
Y_norm=normalization(y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
y=y.reshape(-1,1)
Y_minmax=scaler.fit_transform(y)
Y_minmax=Y_minmax.flatten()
```

```python
fig,(axis1,axis2)=plt.subplots(ncols=2,figsize=(10,5))
axis1.set_title('manual normalization ')
sns.kdeplot(Y_norm,ax=axis1)
axis2.set_title('after scikitlearn normalization')
sns.kdeplot(Y_minmax,ax=axis2)
```
