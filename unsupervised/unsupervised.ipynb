```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]
```

```python
data=list(zip(x,y))
print(data)
```

```python
inertias=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)
plt.plot(range(1,11),inertias,marker='o')
plt.xlabel('number of cluster')
plt.ylabel('inrtia')
plt.title('k-Menas')
```

```python
kmeansdata=KMeans(n_clusters=2)
kmeansdata.fit(data)
plt.scatter(x,y,c=kmeansdata.labels_)
```

```python
new_point=[[12,15]]
predicted_value=kmeansdata.predict(new_point)
print(f"The new point {new_point[0]} belongs to cluster: {predicted_value[0]}")
```

AGGLOMERATIVE CLUSTERING

```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
```

```python
X = np.array([[1, 2], [1, 4], [1, 0],
              [4, 2], [4, 4], [4, 0]])
```

```python
model=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
```

```python
model.fit(X)
```

```python
label=model.labels_
print(label)
```
