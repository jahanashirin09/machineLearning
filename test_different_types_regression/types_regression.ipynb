LINEAR REGRESSION (with single variable)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data.csv')
```

```python
print(data)
```

```python
data.head()
```

```python
data.tail()
```

```python
X=data.iloc[:,0].values.reshape(-1,1)
y=data.iloc[:,-1]
print(X.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
y_pred=lr.predict(X)
```

```python
plt.scatter(X,y,color='b')
plt.plot(X,y_pred,color='g')
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y,y_pred)
```

```python
lr.score(X,y)
```

```python
lr.score(X,y_pred)
```

LOGISTIC REGRESSION (single variable)

```python
logistic=pd.read_csv('dessert.csv')
```

```python
print(logistic)
```

```python
x_log=logistic.iloc[:,:1].values.reshape(-1,1)
y_log=logistic.iloc[:,-1].values.reshape(-1,1)
```

```python
plt.scatter(x_log,y_log,color='g')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_log,y_log)
```

```python
y_log_pred=lg.predict(x_log)
```

```python
plt.scatter(x_log,y_log_pred,color='r')
```

```python
lg.score(x_log,y_log)
```

```python
lg.score(x_log,y_log_pred)
```
